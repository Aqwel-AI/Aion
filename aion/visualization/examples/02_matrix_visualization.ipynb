# Matrix Visualization — Full Examples

This notebook demonstrates **two-dimensional** plotting utilities from **aion.visualization**. Use these for weight matrices, correlation matrices, confusion matrices, attention maps, and sparsity patterns. All functions return a matplotlib `Figure`; use `show=False` to suppress display and `aion.visualization.utils.save_plot(fig, "path.png")` to save.

## 1. Heatmaps and confusion matrices

**plot_matrix_heatmap** — Generic 2D matrix as a color grid (e.g. weights, scores).  
**plot_confusion_matrix** — Classification confusion matrix with optional class labels.  
**plot_confusion_matrix_normalized** — Same data, normalized by row (recall per class).

```python
from aion.visualization import (
    plot_matrix_heatmap,
    plot_confusion_matrix,
    plot_confusion_matrix_normalized,
)

matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
fig = plot_matrix_heatmap(matrix, title="Matrix heatmap", show=True)
```

```python
cm = [[50, 5, 2], [8, 37, 3], [1, 4, 40]]
labels = ["Class A", "Class B", "Class C"]
fig = plot_confusion_matrix(cm, labels=labels, title="Confusion matrix", show=True)
```

```python
fig = plot_confusion_matrix_normalized(cm, labels=labels, title="Normalized (recall)", show=True)
```

## 2. 3D surface and contour

**plot_matrix_surface** — 3D surface plot of matrix values (row, col, value).  
**plot_matrix_contour** — 2D contour map of the same data.

```python
from aion.visualization import plot_matrix_surface, plot_matrix_contour

import numpy as np
grid = np.random.rand(8, 8).tolist()
fig = plot_matrix_surface(grid, title="3D surface", show=True)
```

```python
fig = plot_matrix_contour(grid, title="Contour", show=True)
```

## 3. Annotated heatmap and derived matrices

**plot_matrix_with_values** — Heatmap with numeric annotations in each cell.  
**plot_correlation_matrix** — Computes and plots Pearson correlation from 2D data (rows = samples, columns = features).  
**plot_similarity_matrix** — Similarity matrix from row vectors (metric: "cosine" or "dot").

```python
from aion.visualization import (
    plot_matrix_with_values,
    plot_correlation_matrix,
    plot_similarity_matrix,
)

small = [[1, 2], [3, 4]]
fig = plot_matrix_with_values(small, title="Heatmap with values", show=True)
```

```python
data_2d = np.random.randn(20, 4).tolist()  # 20 samples, 4 features
fig = plot_correlation_matrix(data_2d, labels=["F1", "F2", "F3", "F4"], show=True)
```

```python
vectors = np.random.rand(5, 10).tolist()  # 5 vectors, dim 10
fig = plot_similarity_matrix(vectors, metric="cosine", title="Cosine similarity", show=True)
```

## 4. Matrix histogram and masked heatmap

**plot_matrix_histogram** — Histogram of all matrix values (flattened).  
**plot_masked_heatmap** — Heatmap with a boolean mask (True = hidden/masked).

```python
from aion.visualization import plot_matrix_histogram, plot_masked_heatmap

fig = plot_matrix_histogram(grid, bins=15, title="Matrix value distribution", show=True)
```

```python
mask = [[False, True], [True, False]]  # mask shape must match matrix
fig = plot_masked_heatmap(small, mask, title="Masked heatmap", show=True)
```

## 5. Attention map and sparsity

**plot_attention_map** — Attention weights (e.g. query–key) with optional token labels.  
**plot_matrix_sparsity** — Non-zero pattern (spy plot) for sparse matrices.

```python
from aion.visualization import plot_attention_map, plot_matrix_sparsity

attention = np.random.rand(4, 4).tolist()
tokens = ["A", "B", "C", "D"]
fig = plot_attention_map(attention, tokens=tokens, title="Attention map", show=True)
```

```python
sparse = [[1, 0, 0], [0, 1, 0], [0, 0, 1]]
fig = plot_matrix_sparsity(sparse, title="Sparsity pattern", show=True)
```
