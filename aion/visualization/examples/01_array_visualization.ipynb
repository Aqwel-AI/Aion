# Array Visualization — Full Examples

This notebook demonstrates **one-dimensional** plotting utilities from **aion.visualization**. All functions return a matplotlib `Figure` and support optional display. Use `show=False` to avoid displaying (e.g. in scripts); use `aion.visualization.utils.save_plot(fig, "path.png")` to save figures for reports or papers.

## 1. Basic plots

**plot_array** — Line chart of a 1D sequence (index vs value).  
**plot_histogram** — Distribution of values with configurable bins.  
**plot_scatter** — Relationship between two variables (x vs y).

```python
from aion.visualization import plot_array, plot_histogram, plot_scatter

data = [1.2, 2.1, 1.8, 3.2, 2.9, 4.1, 3.5]
fig1 = plot_array(data, title="Simple 1D array", show=True)
```

```python
fig2 = plot_histogram(data, bins=5, title="Value distribution", show=True)
```

```python
x_vals = [1, 2, 3, 4, 5]
y_vals = [2, 4, 3, 5, 6]
fig3 = plot_scatter(x_vals, y_vals, title="Scatter: x vs y", show=True)
```

## 2. Multiple series and running statistics

**plot_multiple_arrays** — Several curves on one axes (e.g. loss vs val_loss).  
**plot_array_with_mean** — Single series with a horizontal line at the mean.  
**plot_running_mean** — Smooth curve using a sliding-window mean.

```python
from aion.visualization import plot_multiple_arrays, plot_array_with_mean, plot_running_mean

loss = [1.0, 0.7, 0.5, 0.35, 0.25]
val_loss = [1.1, 0.8, 0.55, 0.4, 0.3]
fig = plot_multiple_arrays(
    arrays=[loss, val_loss],
    labels=["train", "val"],
    title="Train vs validation loss",
    show=True
)
```

```python
fig = plot_array_with_mean([10, 12, 9, 11, 10, 13], title="Array with mean", show=True)
```

```python
fig = plot_running_mean(
    [15, 16, 14, 17, 18, 20, 19, 21, 22, 20, 18, 17],
    window_size=4,
    title="Running mean (window=4)",
    show=True
)
```

## 3. Distributions: boxplot, density, CDF

**plot_boxplot** — Box-and-whisker summary of a 1D array.  
**plot_density** — Normalized histogram as a density curve.  
**plot_cdf** — Cumulative distribution function.

```python
from aion.visualization import plot_boxplot, plot_density, plot_cdf

import numpy as np
samples = list(np.random.randn(100) * 2 + 5)

fig = plot_boxplot(samples, title="Boxplot", show=True)
```

```python
fig = plot_density(samples, bins=20, title="Density", show=True)
```

```python
fig = plot_cdf(samples, title="CDF", show=True)
```

## 4. Uncertainty and rolling statistics

**plot_error_bars** — Mean curve with error bars (e.g. ±std).  
**plot_rolling_std** — Rolling standard deviation over a 1D array.  
**plot_min_max_band** — Band between per-index min and max across runs.

```python
from aion.visualization import plot_error_bars, plot_rolling_std, plot_min_max_band

x_pts = [0, 1, 2, 3, 4]
mean_vals = [1.0, 0.8, 0.6, 0.5, 0.4]
err_vals = [0.1, 0.12, 0.08, 0.07, 0.06]
fig = plot_error_bars(x_pts, mean_vals, err_vals, title="Mean ± error", show=True)
```

```python
series = list(np.random.randn(50).cumsum() + 10)
fig = plot_rolling_std(series, window_size=5, title="Rolling std (window=5)", show=True)
```

```python
series_for_band = [1.0, 0.9, 0.7, 0.5, 0.4, 0.35, 0.3, 0.28, 0.25]
fig = plot_min_max_band(series_for_band, window_size=3, title="Min–max band (window=3)", show=True)
```

## 5. Advanced: autocorrelation, quantiles, scatter with fit, dual axis

**plot_autocorrelation** — Autocorrelation of a 1D signal (e.g. time series).  
**plot_quantiles** — Quantile levels over the array.  
**plot_scatter_with_fit** — Scatter plot with a linear fit line.  
**plot_dual_axis** — Two series on left/right y-axes.

```python
from aion.visualization import (
    plot_autocorrelation,
    plot_quantiles,
    plot_scatter_with_fit,
    plot_dual_axis,
)

signal = list(np.sin(np.linspace(0, 4 * np.pi, 100)) + np.random.randn(100) * 0.2)
fig = plot_autocorrelation(signal, title="Autocorrelation", show=True)
```

```python
fig = plot_quantiles(signal, qs=[0.25, 0.5, 0.75], title="Quantiles", show=True)
```

```python
x = list(np.linspace(0, 10, 20))
y = [xi * 1.5 + 2 + np.random.randn() * 0.5 for xi in x]
fig = plot_scatter_with_fit(x, y, title="Scatter with linear fit", show=True)
```

```python
x_axis = [1, 2, 3, 4, 5]
y1_vals = [1, 2, 3, 4, 5]
y2_vals = [100, 80, 60, 40, 20]
fig = plot_dual_axis(
    x_axis,
    y1_vals,
    y2_vals,
    label1="Metric A",
    label2="Metric B",
    title="Dual axis",
    show=True
)
```
