# Training Visualization — Full Examples

This notebook demonstrates **training and experiment** plotting from **aion.visualization**. Use these to visualize loss, accuracy, learning rate, and other metrics over epochs or steps. All functions return a matplotlib `Figure`; use `show=False` when scripting and `save_plot(fig, "path.png")` to export.

## 1. Training history and single metric

**plot_training_history** — Plots all metrics in a dictionary (e.g. loss, val_loss, accuracy) on one axes.  
**plot_metric** — Plots one metric from that history dictionary.

```python
from aion.visualization import plot_training_history, plot_metric

history = {
    "loss": [1.0, 0.7, 0.5, 0.35, 0.25, 0.18],
    "val_loss": [1.1, 0.8, 0.55, 0.4, 0.32, 0.28],
    "accuracy": [0.5, 0.65, 0.75, 0.82, 0.87, 0.90],
}
fig = plot_training_history(history, show=True)
```

```python
fig = plot_metric(history, "accuracy", title="Accuracy over epochs", show=True)
```

## 2. Train vs val and learning rate

**plot_train_vs_val** — Two sequences (train and validation) on the same axes.  
**plot_learning_rate** — Learning rate schedule over steps or epochs.

```python
from aion.visualization import plot_train_vs_val, plot_learning_rate

train = history["loss"]
val = history["val_loss"]
fig = plot_train_vs_val(train, val, title="Train vs validation loss", show=True)
```

```python
lr_schedule = [1e-3, 8e-4, 6e-4, 4e-4, 2e-4, 1e-4]
fig = plot_learning_rate(lr_schedule, title="Learning rate schedule", show=True)
```

## 3. Best value and metrics grid

**plot_metric_with_best** — Plots one metric and highlights the best value (mode: "min" or "max").  
**plot_metrics_grid** — Multiple metrics in a grid of subplots.

```python
from aion.visualization import plot_metric_with_best, plot_metrics_grid

fig = plot_metric_with_best(
    history,
    "val_loss",
    mode="min",
    title="Val loss (best = min)",
    show=True
)
```

```python
fig = plot_metrics_grid(history, cols=2, title="Metrics grid", show=True)
```

## 4. Confidence band and early stopping

**plot_confidence_band** — Mean curve with ±std shaded band (e.g. across runs).  
**plot_early_stopping** — Plots a metric and marks where early stopping would trigger (patience, mode).

```python
from aion.visualization import plot_confidence_band, plot_early_stopping

mean_vals = [1.0, 0.7, 0.5, 0.35, 0.25, 0.2]
std_vals = [0.1, 0.08, 0.06, 0.05, 0.04, 0.03]
fig = plot_confidence_band(mean_vals, std_vals, title="Mean ± std", show=True)
```

```python
fig = plot_early_stopping(
    history,
    "val_loss",
    patience=2,
    mode="min",
    title="Early stopping (patience=2)",
    show=True
)
```

## 5. Epoch time

**plot_epoch_time** — Plot per-epoch duration (e.g. seconds). Useful for profiling and scaling.

```python
from aion.visualization import plot_epoch_time

epoch_times = [12.1, 11.8, 11.9, 12.0, 11.7, 11.8]  # seconds
fig = plot_epoch_time(epoch_times, title="Epoch duration", show=True)
```
